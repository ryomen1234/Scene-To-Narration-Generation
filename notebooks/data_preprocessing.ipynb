```python
raw_data_path = Path("data/raw")
print(raw_data_path.exists())
```

```python
Images = raw_data_path / "Images"
Captions = raw_data_path / "captions.txt"

print(f"Image paths: {Images.exists()}")
print(f"Captions path: {Captions.exists()}")
```

```python
from collections import defaultdict

def change_format(cap_file: Path):

    caption_map = defaultdict(list)

    with  open(cap_file, "r") as f:
        next(f)
        for line in f:
            img, cap = line.strip().split(",",1)
            caption_map[img].append(cap)
    
    return caption_map
            
```

```python
caption_map = change_format(Captions)
caption_map
```

```python
import random 

images = list(caption_map.keys())
random.shuffle(images)

sequence = [images[i:i+5] for i in range(0, len(images), 5)]
```

```python
dataset = []

for seq in sequence:

    if len(seq) < 5:
        continue
    
    caps = []

    for img in seq:
         cap = random.choice(caption_map[img])
         caps.append(cap)
    
    story = " ".join(caps)

    dataset.append({
        "images" : [Images / img for img in seq],
        "story": story
    })
```
